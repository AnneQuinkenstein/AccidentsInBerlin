```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../data/GeneralDatensatz18-21.csv", sep=';', encoding='ISO-8859-1')
```

```python
font = {'family' : 'serif',
        'size'   : 28}

plt.rc('font', **font)

sns.set(style="darkgrid", color_codes=True)
sns.set_context("paper", font_scale=1.5)
```

```python
fig, ax = plt.subplots(5, 2, figsize=(20, 10))
for i, val in enumerate(['UJAHR', 'UMONAT', 'UWOCHENTAG', 'UKATEGORIE', 'UART', 'UTYP1', 'USTRZUSTAND', 'IstRad', 'IstFuss', 'IstPKW']):
    row = i // 2  # Correct row calculation
    col = i % 2   # Correct column calculation
    data.hist(column=val, ax=ax[row, col], bins=20)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.savefig("./daten_histogramme.png")
```
