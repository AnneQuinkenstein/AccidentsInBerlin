```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.metrics import make_scorer, fbeta_score

```

```python
train_data_loaded = pd.read_csv('../data/train_data_2024-08-01.csv')

X = train_data_loaded.drop(columns=['UKATEGORIE'])
y = train_data_loaded['UKATEGORIE']

# KFold-Konfiguration
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Definieren des F-beta-Scores mit beta = 2
beta = 2
fbeta_scorer = make_scorer(fbeta_score, beta=beta)
```

```python
from sklearn.linear_model import LogisticRegression

param_dist = {
    'C': [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.1, 0.25, 0.5, 0.75, 1],
    'max_iter': list(range(500, 5000, 100)),
    'tol': [1e-4, 1e-3, 1e-2],
    'penalty': ['l2', 'l1', 'elasticnet'],
    'solver': ['lbfgs', 'liblinear', 'saga'],    
    
}

stratified_kfold = StratifiedKFold(n_splits=5)
model = LogisticRegression(random_state=42, class_weight = {0:1, 1: 9})
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=100, cv=stratified_kfold, scoring=fbeta_scorer)
random_search.fit(X, y)
best_params = random_search.best_params_
```
