```python
import pandas as pd
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import  cross_val_score, StratifiedKFold
from sklearn.metrics import fbeta_score, make_scorer
import matplotlib.pyplot as plt

train_data_loaded = pd.read_csv('../data/train_data_2024-08-01.csv')

X = train_data_loaded.drop(columns=['UKATEGORIE'])
y = train_data_loaded['UKATEGORIE']

# KFold-Konfiguration
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Definieren des F-beta-Scores mit beta = 2
beta = 2
fbeta_scorer = make_scorer(fbeta_score, beta=beta)

##Parameter für Hyperparametertuning
n_estimators = list(range(100, 3000, 20))
max_depth = list(range(10, 500, 10))
min_samples_split = list(range(2, 500, 4))
min_samples_leaf = list(range(2, 100, 3))
max_features =  ['sqrt', 'log2',2, 4, 6, 8, 10, 12, 14, 16, 18]

```

```python
estimators = []

for value in n_estimators:
    model = BalancedRandomForestClassifier(n_estimators=value, random_state=42, replacement=True, bootstrap=False)
    fbeta = cross_val_score(model, X, y, cv=kf, scoring=fbeta_scorer)
    print(f"n_estimators: {value} Fbeta Score Bal RF (k-fold): {fbeta.mean()}")
    estimators.append(fbeta.mean())
    

# Fbeta-Scores plotten
plt.figure(figsize=(10, 6))
plt.plot(n_estimators, estimators, marker='o', label='n_estimators BalancedRandomForestClassifier')
plt.title('Fbeta-Scores für verschiedene estimators')
plt.xlabel('n_estimators')
plt.ylabel('Fbeta-Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
depths = []

for value in max_depth:
    model = BalancedRandomForestClassifier(max_depth=value, random_state=42, replacement=True, bootstrap=False)
    fbeta = cross_val_score(model, X, y, cv=kf, scoring=fbeta_scorer)
    print(f"max_depth: {value} Fbeta Score Bal RF (k-fold): {fbeta.mean()}")
    depths.append(fbeta.mean())
    

# Fbeta-Scores plotten
plt.figure(figsize=(10, 6))
plt.plot(max_depth, depths, marker='o', label='n_estimators BalancedRandomForestClassifier')
plt.title('Fbeta-Scores für verschiedene max_depths')
plt.xlabel('n_estimators')
plt.ylabel('Fbeta-Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
splits = []

for value in min_samples_split:
    model = BalancedRandomForestClassifier(min_samples_split=value, random_state=42, replacement=True, bootstrap=False)
    fbeta = cross_val_score(model, X, y, cv=kf, scoring=fbeta_scorer)
    print(f"min_samples_split: {value} Fbeta Score Bal RF (k-fold): {fbeta.mean()}")
    splits.append(fbeta.mean())
    

# Fbeta-Scores plotten
plt.figure(figsize=(10, 6))
plt.plot(min_samples_split, splits, marker='o', label='min_samples_split BalancedRandomForestClassifier')
plt.title('Fbeta-Scores für verschiedene splits')
plt.xlabel('min_samples_split')
plt.ylabel('Fbeta-Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
features = []

for value in max_features:
    model = BalancedRandomForestClassifier(max_features=value, random_state=42, replacement=True, bootstrap=False)
    fbeta = cross_val_score(model, X, y, cv=kf, scoring=fbeta_scorer)
    print(f"max_features: {value} Fbeta Score Bal RF (k-fold): {fbeta.mean()}")
    features.append(fbeta.mean())
    

# Fbeta-Scores plotten
plt.figure(figsize=(10, 6))
plt.plot(max_features, features, marker='o', label='max_features BalancedRandomForestClassifier')
plt.title('Fbeta-Scores für verschiedene max_features')
plt.xlabel('max_features')
plt.ylabel('Fbeta-Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
leaves = []

for value in min_samples_leaf:
    model = BalancedRandomForestClassifier(min_samples_leaf=value, random_state=42, replacement=True, bootstrap=False)
    fbeta = cross_val_score(model, X, y, cv=kf, scoring=fbeta_scorer)
    print(f"min_samples_leaf: {value} Fbeta Score Bal RF (k-fold): {fbeta.mean()}")
    leaves.append(fbeta.mean())
    

# Fbeta-Scores plotten
plt.figure(figsize=(10, 6))
plt.plot(min_samples_leaf, leaves, marker='o', label='min_samples_leaf BalancedRandomForestClassifier')
plt.title('Fbeta-Scores für verschiedene leaves')
plt.xlabel('min_samples_leaf')
plt.ylabel('Fbeta-Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
max_score_estimators= max(estimators)
index_max_score_estimators = estimators.index(max_score_estimators)
print(max_score_estimators)
print(index_max_score_estimators)
```

```python
max_score_depths= max(depths)
index_max_score_depths = depths.index(max_score_depths)
print(max_score_depths)
print(index_max_score_depths)
```
